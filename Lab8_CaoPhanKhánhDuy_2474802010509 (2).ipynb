## Câu 2: Liệt kê danh sách các stopword theo từng ngôn ngữ
Viết chương trình Python với thư viện NLTK để liệt kê danh sách các stopword bằng các ngôn ngữ khác nhau.

```
from nltk.corpus import stopwords

languages = stopwords.fileids()
print("Các ngôn ngữ có sẵn stopword:", languages)

# In ra 10 stopword đầu tiên của một vài ngôn ngữ tiêu biểu
for lang in ['english', 'french', 'spanish', 'german']:
    if lang in languages:
        print(f"\n{lang.upper()} (10 từ đầu):", stopwords.words(lang)[:10])
```

(số lượng stopword của từng ngôn ngữ và kiểm tra 1 từ có phải là stopword hay không).

## Câu 4: Loại bỏ các stopword từ một văn bản đã cho
Viết chương trình Python với thư viện NLTK để loại bỏ các stopword từ một văn bản đã cho.

```
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

text = "This is a sample sentence, showing off the stop words filtration for NLTK analysis."

stop_words = set(stopwords.words('english'))
tokens = word_tokenize(text)

filtered_tokens = [w for w in tokens if w.lower() not in stop_words]

print("Văn bản gốc (tokenized):", tokens)
print("\nVăn bản sau khi loại bỏ stopword:", filtered_tokens)
```

## Câu 5: Bỏ qua (lọc) các stopword từ danh sách các từ cho trước
Viết chương trình Python với thư viện NLTK để bỏ qua các stopword từ danh sách các stopword
(áp dụng trên một danh sách từ, không cần tokenize câu văn).

```
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))

word_list = ['I', 'am', 'learning', 'Natural', 'Language', 'Processing',
             'with', 'the', 'NLTK', 'library', 'and', 'it', 'is', 'very', 'interesting']

filtered_list = [w for w in word_list if w.lower() not in stop_words]

print("Danh sách từ gốc:", word_list)
print("\nDanh sách sau khi bỏ qua stopword:", filtered_list)
```

## Câu 6: Tìm định nghĩa và ví dụ của một từ bằng WordNet
Viết chương trình Python với thư viện NLTK để tìm định nghĩa và ví dụ của một từ đã cho bằng WordNet.

```
from nltk.corpus import wordnet

word = "happy"

synsets = wordnet.synsets(word)
print(f"Số lượng synset (nghĩa) của từ '{word}': {len(synsets)}\n")

for syn in synsets:
    print(f"- Synset: {syn.name()}")
    print(f"  Định nghĩa: {syn.definition()}")
    print(f"  Ví dụ: {syn.examples()}")
    print()
```

## Câu 7: Tìm tập từ đồng nghĩa và trái nghĩa của một từ
Viết chương trình Python với thư viện NLTK để tìm tập hợp các từ đồng nghĩa và trái nghĩa của một từ nào đó.

```
from nltk.corpus import wordnet

word = "good"
synonyms = []
antonyms = []

for syn in wordnet.synsets(word):
    for lemma in syn.lemmas():
        synonyms.append(lemma.name())
        if lemma.antonyms():
            antonyms.append(lemma.antonyms()[0].name())

print(f"Từ đồng nghĩa của '{word}':", set(synonyms))
print(f"\nTừ trái nghĩa của '{word}':", set(antonyms))
```

## Câu 9: So sánh sự giống nhau (similarity) giữa hai danh từ
Viết chương trình Python với thư viện NLTK để so sánh sự giống nhau của hai danh từ đã cho.

```
from nltk.corpus import wordnet

noun1 = wordnet.synsets('car', pos=wordnet.NOUN)[0]
noun2 = wordnet.synsets('automobile', pos=wordnet.NOUN)[0]
noun3 = wordnet.synsets('banana', pos=wordnet.NOUN)[0]

print(f"'{noun1.name()}' vs '{noun2.name()}'")
print("  Path similarity:", noun1.path_similarity(noun2))
print("  Wu-Palmer similarity:", noun1.wup_similarity(noun2))

print(f"\n'{noun1.name()}' vs '{noun3.name()}'")
print("  Path similarity:", noun1.path_similarity(noun3))
print("  Wu-Palmer similarity:", noun1.wup_similarity(noun3))
```

## Câu 11: Đếm số lượng tên nam và nữ trong corpus 'names'
Viết chương trình Python với thư viện NLTK để tìm số lượng tên nam và nữ trong kho ngữ liệu tên (names corpus).
In ra 10 tên nam và 10 tên nữ đầu tiên.

> Ghi chú: Kho ngữ liệu tên chứa tổng cộng khoảng 2943 tên nam (male.txt) và 5001 tên nữ (female.txt),
> được biên soạn bởi Kantrowitz, Ross.

## Câu 12: In 15 kết hợp ngẫu nhiên tên đã được gắn nhãn nam/nữ
Viết chương trình Python với thư viện NLTK để in 15 kết hợp ngẫu nhiên đầu tiên được gắn nhãn nam
và được gắn nhãn tên nữ từ kho tên.

```
import random
from nltk.corpus import names

male_names = names.words('male.txt')
female_names = names.words('female.txt')

labeled_names = ([(name, 'male') for name in male_names] +
                  [(name, 'female') for name in female_names])

random.shuffle(labeled_names)

print("15 kết hợp (tên, nhãn) ngẫu nhiên đầu tiên:")
for item in labeled_names[:15]:
    print(item)
```

## Câu 13: Trích xuất ký tự cuối cùng của tên và tạo mảng đặc trưng mới
Viết chương trình Python với thư viện NLTK để trích xuất ký tự cuối cùng của tất cả các tên được gắn nhãn
và tạo mảng mới với chữ cái cuối cùng của mỗi tên cùng với nhãn được liên kết.

```
# Sử dụng lại danh sách labeled_names đã được xáo trộn ở câu 12
featuresets = [(name[-1].lower(), gender) for (name, gender) in labeled_names]

print("15 phần tử đầu tiên của featuresets (ký tự cuối, nhãn):")
for item in featuresets[:15]:
    print(item)

print(f"\nTổng số phần tử trong featuresets: {len(featuresets)}")
```
